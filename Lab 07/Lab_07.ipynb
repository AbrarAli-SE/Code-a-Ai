## Task 1: String Evolution with Different Strings

```python
import random

# GENES = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ 1234567890!@#$%^&*()"
GENES = '''abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ 1234567890,.-;:_!"#%&/()=?@${[]}'''

class Individual:
    def __init__(self, chromosome):
        self.chromosome = chromosome
        self.fitness = self.cal_fitness()

    @classmethod
    def mutated_genes(cls):
        return random.choice(GENES)

    @classmethod
    def create_gnome(cls, target):
        return ''.join([cls.mutated_genes() for _ in range(len(target))])

    def mate(self, par2):
        child_chromosome = []
        for gp1, gp2 in zip(self.chromosome, par2.chromosome):
            prob = random.random()
            if prob < 0.45:
                child_chromosome.append(gp1)
            elif prob < 0.90:
                child_chromosome.append(gp2)
            else:
                child_chromosome.append(self.mutated_genes())
        return Individual(''.join(child_chromosome))

    def cal_fitness(self):
        global TARGET
        fitness = sum(1 for gs, gt in zip(self.chromosome, TARGET) if gs != gt)
        return fitness

def genetic_algorithm(target, population_size=100):
    global TARGET
    TARGET = target
    generation = 1
    found = False
    population = [Individual(Individual.create_gnome(target)) for _ in range(population_size)]

    while not found:
        population = sorted(population, key=lambda x: x.fitness)

        if population[0].fitness == 0:
            found = True
            break

        new_generation = population[:10]  # Elitism

        for _ in range(population_size - 10):
            parent1 = random.choice(population[:50])
            parent2 = random.choice(population[:50])
            child = parent1.mate(parent2)
            new_generation.append(child)

        population = new_generation

        print(f"Generation: {generation}\tString: {population[0].chromosome}\tFitness: {population[0].fitness}")
        generation += 1

    print(f"\nTarget: {target}\nGeneration: {generation}\nString: {population[0].chromosome}")

# Test with different strings
genetic_algorithm("Hello World")
print("\n" + "="*50 + "\n")
genetic_algorithm("AI Lab 2025")
```

```python
# Task 2.2: Different Crossover & Mutation Probabilities
class Individual_Custom:
    def __init__(self, chromosome):
        self.chromosome = chromosome
        self.fitness = self.cal_fitness()

    @classmethod
    def mutated_genes(cls):
        return random.choice(GENES)

    @classmethod
    def create_gnome(cls, target):
        return ''.join([cls.mutated_genes() for _ in range(len(target))])

    def mate(self, par2, crossover_prob=0.9, mutation_prob=0.1):
        child_chromosome = []
        for gp1, gp2 in zip(self.chromosome, par2.chromosome):
            prob = random.random()
            if prob < crossover_prob/2:
                child_chromosome.append(gp1)
            elif prob < crossover_prob:
                child_chromosome.append(gp2)
            else:
                if random.random() < mutation_prob:
                    child_chromosome.append(self.mutated_genes())
                else:
                    child_chromosome.append(gp1)
        return Individual_Custom(''.join(child_chromosome))

    def cal_fitness(self):
        global TARGET
        return sum(1 for gs, gt in zip(self.chromosome, TARGET) if gs != gt)

def ga_custom(target, crossover_prob=0.9, mutation_prob=0.1):
    global TARGET
    TARGET = target
    generation = 1
    population = [Individual_Custom(Individual_Custom.create_gnome(target)) for _ in range(100)]

    while True:
        population = sorted(population, key=lambda x: x.fitness)
        if population[0].fitness == 0:
            break

        new_generation = population[:10]
        for _ in range(90):
            parent1 = random.choice(population[:50])
            parent2 = random.choice(population[:50])
            child = parent1.mate(parent2, crossover_prob, mutation_prob)
            new_generation.append(child)
        population = new_generation
        generation += 1

    print(f"Crossover: {crossover_prob}, Mutation: {mutation_prob} -> Generations: {generation}")

# Test different probabilities
print("\nANALYZING CROSSOVER & MUTATION PROBABILITIES\n")
for cp in [0.7, 0.8, 0.9]:
    for mp in [0.05, 0.1, 0.2]:
        ga_custom("Python", cp, mp)
```

## Task 3: N-Queens Problem

```python
import random

class NQueens:
    def __init__(self, n=8):
        self.n = n
        self.chromosome = list(range(n))
        random.shuffle(self.chromosome)
        self.fitness = self.calculate_fitness()

    def calculate_fitness(self):
        clashes = 0
        for i in range(self.n):
            for j in range(i + 1, self.n):
                if abs(self.chromosome[i] - self.chromosome[j]) == abs(i - j):
                    clashes += 1
        return clashes

    def mate(self, partner):
        crossover_point = random.randint(1, self.n - 1)
        child_chromosome = self.chromosome[:crossover_point] + partner.chromosome[crossover_point:]

        child = NQueens(self.n)
        child.chromosome = child_chromosome

        # Mutation
        if random.random() < 0.2:
            i, j = random.sample(range(self.n), 2)
            child.chromosome[i], child.chromosome[j] = child.chromosome[j], child.chromosome[i]

        child.fitness = child.calculate_fitness()
        return child

def solve_n_queens(n=8, population_size=100, max_generations=10000):
    population = [NQueens(n) for _ in range(population_size)]
    generation = 0

    while generation < max_generations:
        population = sorted(population, key=lambda x: x.fitness)

        if population[0].fitness == 0:
            print(f"\n✓ Solution Found!")
            print(f"Generation: {generation}")
            print(f"Board Configuration: {population[0].chromosome}")
            print_board(population[0].chromosome)
            return population[0].chromosome

        if generation % 100 == 0:
            print(f"Gen {generation}: Best Fitness = {population[0].fitness}")

        # New generation
        new_population = population[:20]  # Elitism

        while len(new_population) < population_size:
            parent1 = random.choice(population[:50])
            parent2 = random.choice(population[:50])
            child = parent1.mate(parent2)
            new_population.append(child)

        population = new_population
        generation += 1

    print(f"\nBest solution after {max_generations} generations:")
    print(f"Fitness (clashes): {population[0].fitness}")
    print(f"Configuration: {population[0].chromosome}")
    return population[0].chromosome

def print_board(solution):
    n = len(solution)
    print("\nChessboard:")
    for i in range(n):
        row = ['. ' for _ in range(n)]
        row[solution[i]] = 'Q '
        print(''.join(row))

# Solve N-Queens
print("="*50)
print("N-QUEENS PROBLEM USING GENETIC ALGORITHM")
print("="*50)
solve_n_queens(8, 100)

# Test with different N values
print("\n\n--- Solving 4-Queens ---")
solve_n_queens(4, 50)
```
