## Setup & Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better visualizations
plt.style.use('ggplot')
%matplotlib inline
```

## Task 1: Fruits Price Analysis & Bar Graph

```python
# Task 1: Fruits Dictionary & Bar Graph
data = {'Fruits': ['Mango', 'Apple', 'Banana'],
        'Rwp': [80, 40, 60],
        'Lhr': [100, 30, 90],
        'Isb': [90, 50, 100]}

df_fruits = pd.DataFrame(data)
print("Fruits Price DataFrame:")
print(df_fruits)

# Set Fruits as index for better plotting
df_fruits.set_index('Fruits', inplace=True)

# Bar Graph
df_fruits.plot(kind='bar', figsize=(10, 6), width=0.8)
plt.title('Fruit Prices Across Cities', fontsize=16, fontweight='bold')
plt.xlabel('Fruits', fontsize=12)
plt.ylabel('Price (PKR)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Cities')
plt.tight_layout()
plt.show()

# Grouped Bar Chart (Alternative)
df_fruits.T.plot(kind='bar', figsize=(10, 6))
plt.title('City-wise Fruit Prices', fontsize=16, fontweight='bold')
plt.xlabel('Cities', fontsize=12)
plt.ylabel('Price (PKR)', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Fruits')
plt.tight_layout()
plt.show()
```

## Task 2: Student Scores Dataset Analysis

```python
# Task 2: Create Student Dataset
students_data = {
    'Name': ['Ali', 'Sara', 'Ahmed', 'Fatima', 'Hassan', 'Ayesha', 'Usman', 'Zara', 'Bilal', 'Hina'],
    'Math': [85, 92, 78, 95, 88, 76, 90, 82, 79, 94],
    'Science': [78, 88, 85, 90, 82, 80, 87, 91, 75, 89],
    'English': [92, 85, 88, 87, 90, 82, 84, 86, 91, 88],
    'Computer': [88, 90, 82, 92, 85, 78, 91, 87, 84, 93],
    'City': ['Rwp', 'Isb', 'Lhr', 'Rwp', 'Isb', 'Lhr', 'Rwp', 'Isb', 'Lhr', 'Rwp']
}

df_students = pd.DataFrame(students_data)

# Display first few rows
print("=" * 60)
print("FIRST FEW ROWS:")
print("=" * 60)
print(df_students.head())

# Summary Statistics
print("\n" + "=" * 60)
print("SUMMARY STATISTICS:")
print("=" * 60)
print(df_students.describe())

# DataFrame Info
print("\n" + "=" * 60)
print("DATAFRAME INFO:")
print("=" * 60)
print(df_students.info())

# Check for Missing Values
print("\n" + "=" * 60)
print("MISSING VALUES:")
print("=" * 60)
print(df_students.isnull().sum())

# DATA ANALYSIS
print("\n" + "=" * 60)
print("DATA ANALYSIS:")
print("=" * 60)

# Average score for each subject
subjects = ['Math', 'Science', 'English', 'Computer']
avg_scores = df_students[subjects].mean()
print("\nAverage Scores by Subject:")
print(avg_scores)

# Highest and Lowest scores for each subject
print("\nHighest Scores:")
print(df_students[subjects].max())
print("\nLowest Scores:")
print(df_students[subjects].min())

# Overall Average
df_students['Average'] = df_students[subjects].mean(axis=1)
print("\nStudent-wise Average:")
print(df_students[['Name', 'Average']])

print(f"\nOverall Class Average: {df_students['Average'].mean():.2f}")
```

## Task 2: Visualizations

```python
# VISUALIZATION 1: Histogram - Distribution of Scores
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Scores by Subject', fontsize=16, fontweight='bold')

for idx, subject in enumerate(subjects):
    row = idx // 2
    col = idx % 2
    axes[row, col].hist(df_students[subject], bins=8, color=plt.cm.Set3(idx), edgecolor='black', alpha=0.7)
    axes[row, col].set_title(f'{subject} Scores Distribution')
    axes[row, col].set_xlabel('Scores')
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# VISUALIZATION 2: Bar Plot - Average Scores by Subject
plt.figure(figsize=(10, 6))
avg_scores.plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'], edgecolor='black')
plt.title('Average Scores by Subject', fontsize=16, fontweight='bold')
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Average Score', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 100)
for i, v in enumerate(avg_scores):
    plt.text(i, v + 1, f'{v:.1f}', ha='center', fontweight='bold')
plt.tight_layout()
plt.show()

# VISUALIZATION 3: Pie Chart - Distribution of Average Scores
plt.figure(figsize=(10, 8))
avg_scores.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                colors=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'])
plt.title('Distribution of Average Scores Across Subjects', fontsize=16, fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()

# VISUALIZATION 4: Box Plot - Score Distribution
plt.figure(figsize=(10, 6))
df_students[subjects].boxplot(patch_artist=True)
plt.title('Score Distribution by Subject (Box Plot)', fontsize=16, fontweight='bold')
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# VISUALIZATION 5: Heatmap - Student Performance
plt.figure(figsize=(10, 8))
sns.heatmap(df_students[subjects].T, annot=True, fmt='d', cmap='YlOrRd',
            xticklabels=df_students['Name'], yticklabels=subjects, linewidths=0.5)
plt.title('Student Performance Heatmap', fontsize=16, fontweight='bold')
plt.xlabel('Students', fontsize=12)
plt.ylabel('Subjects', fontsize=12)
plt.tight_layout()
plt.show()
```

## Task 3: Insert New Column in DataFrame

```python
# Task 3: Insert New Column
print("=" * 60)
print("INSERTING NEW COLUMNS:")
print("=" * 60)

# Method 1: Simple Assignment
df_students['Grade'] = df_students['Average'].apply(
    lambda x: 'A+' if x >= 90 else 'A' if x >= 80 else 'B' if x >= 70 else 'C'
)

# Method 2: Using insert() - Insert at specific position
df_students.insert(2, 'Roll_No', range(101, 101 + len(df_students)))

# Method 3: Add Total Marks
df_students['Total'] = df_students[subjects].sum(axis=1)

# Method 4: Add Percentage
df_students['Percentage'] = (df_students['Total'] / 400) * 100

print("\nDataFrame after inserting new columns:")
print(df_students)

# Display specific columns
print("\n" + "=" * 60)
print("STUDENT REPORT CARD:")
print("=" * 60)
print(df_students[['Name', 'Roll_No', 'Average', 'Grade', 'Percentage']])
```
